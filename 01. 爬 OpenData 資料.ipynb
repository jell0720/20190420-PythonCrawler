## 接下來搜尋自己想要的資訊

### 比如我想知道 文化中心站 就在 filter 內下達:  StationID eq 'O7' 來過濾條件



### 然後你就會取得一個 url 可以放到程式內使用:

<h3>
<pre>
https://ptx.transportdata.tw/MOTC/v2/Rail/Metro/LiveBoard/KRTC?$filter=StationID%20eq%20'O7'&$top=30&$format=JSON
